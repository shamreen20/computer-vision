# Harris Corner Detection
cornerHarris Function

src Input single-channel 8-bit or floating-point image.

dst Image to store the Harris detector responses. It has the type CV_32FC1 and the same size as src .

blockSize Neighborhood size (see the details on #cornerEigenValsAndVecs ).

ksize Aperture parameter for the Sobel operator.

k Harris detector free parameter. See the formula in DocString

borderType Pixel extrapolation method. See #BorderTypes.

```python
flat_chess = cv2.imread('C:\\Users\\Rafiun Nesha\\Desktop\\images\\download6.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,7))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
#display_img(img)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
image = cv2.imread("C:\\Users\\Rafiun Nesha\\Desktop\\images\\download7.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display_img(image)
```

```python
image_gray = cv2.imread("C:\\Users\\Rafiun Nesha\\Desktop\\images\\download7.png",0)
display_img(image_gray)
```

```python
img_gray.shape
```

```python
# Convert Gray Scale Image to Float Values
gray = np.float32(gray_flat_chess)

# Corner Harris Detection
Harris = cv2.cornerHarris(src=gray,blockSize=2, ksize=3, k=0.04)

# result is dilated for marking the corners, not important to actual corner detection
# this is just so we can plot out the points on the image shown
Harris = cv2.dilate(Harris,None)

# Threshold for an optimal value, it may vary depending on the image.
flat_chess[Harris > 0.01*Harris.max()]=[218,218,3]

display_img(img)
```

```python
# Convert Gray Scale Image to Float Values
gray = np.float32(image_gray)

# Corner Harris Detection
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)

# result is dilated for marking the corners, not important to actual corner detection
# this is just so we can plot out the points on the image shown
dst = cv2.dilate(dst,None)

# Threshold for an optimal value, it may vary depending on the image.
image[dst>0.01*dst.max()]=[255,0,0]

display_img(image)
```

# Shi-Tomasi Corner Detector & Good Features to Track Paper

```python
img = cv2.imread("C:\\Users\\Rafiun Nesha\\Desktop\\images\\download6.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
display_img(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display_img(img_gray)
```

```python
corners = cv2.goodFeaturesToTrack(img_gray,64,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),3,255,-1)
display_img(img)
```
