#**Canny edge detection is a multi-step algorithm that can detect edges with noise supressed at the same time. Smooth the image with a Gaussian filter to reduce noise and unwanted details and textures.**

#Import import libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread("C:\\Users\\Rafiun Nesha\\Desktop\\images\\downloadDogFace.jpg")
plt.imshow(image)
```

```python
image_edge = cv2.Canny(image=image, threshold1=120, threshold2=120)
plt.imshow(image_edge)
```

```python
image1 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image1_edge = cv2.Canny(image=image1, threshold1=0, threshold2=255)
plt.imshow(image1_edge)
```

#**We see here the above two images, there are lots of noises. So, for Reducing the noises we use suitable threshold values for image.**

# Choose the most appropriate threshold values

```python
# Calculate the median pixel value
med_val = np.median(image1)
```

```python
# Lower bound is either 0 or 70% of the median value, whicever is higher
lower = int(max(0, 0.7*med_val))
```

```python
# Upper bound is either 255 or 30% above the median value, whichever is lower
upper = int(min(255,1.4*med_val))
```

```python
edges = cv2.Canny(image=image, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

**Sometimes it helps to blur the images first, so we don't pick up minor edges.**

```python
blurred_img = cv2.blur(image,ksize=(5,5))
```

```python
img_edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper)
```

```python
plt.imshow(img_edges)
```
