#Grid Detection
 We've only scratched the surface here :)

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
chess_board = cv2.imread("C:\\Users\\Rafiun Nesha\\Desktop\\images\\download6.png")
plt.imshow(chess_board)
```

```python
found, corners = cv2.findChessboardCorners(chess_board,(7,7))
```

```python
if found:
    print('OpenCV is able to find the corners')
else:
    print('OpenCV did not find corners. Double check your patternSize.')
```

```python
corners.shape
```

```python
chess_board_copy = chess_board.copy()
cv2.drawChessboardCorners(chess_board_copy, (7,7), corners, found)
```

```python
plt.imshow(chess_board_copy)
```

#Circle Grid

```python
circle_dots = cv2.imread("C:\\Users\\Rafiun Nesha\\Desktop\\images\\download8.png")
plt.imshow(circle_dots)
```

```python
found, corners = cv2.findCirclesGrid(circle_dots, (10,10), cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
found
```

```python
circle_dots_copy = circle_dots.copy()
cv2.drawChessboardCorners(circle_dots_copy, (10,10), corners, found)
```
